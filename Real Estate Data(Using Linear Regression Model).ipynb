# Linear Regression Model

```python
x = df1.drop(['House price of unit area'], axis = 1)
```

```python
y = df1['House price of unit area']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()
```

```python
lreg.fit(x_train, y_train)
```

```python
lreg.score(x, y)
```

```python
accuracy = lreg.score(x_test, y_test)
```

```python
print("Accuracy: ", accuracy*100, '%')
```

```python
print(lreg.predict([[2022.117, 5, 7, 6, 25, 121]]))
```

# R Squared Value Calculation

```python
from sklearn.metrics import r2_score
```

```python
predctions = lreg.predict(x_test)
```

```python
r2_score(y_test, predctions)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y_test, predctions, color = 'green')
plt.title('Show Prediction')
plt.xlabel("Real Data")
plt.ylabel("Predicted Data")
```

```python
test_residulas = y_test - predctions
```

```python
sns.displot(test_residulas, bins = 25, kde = True, color = 'green')
```
